```python
import pandas as pd
```

```python
df = pd.read_csv("pickuplineig.csv")
```

```python
combined = ""
for i in df['tweet']:
    i = i.replace("\n\n", " ")
    combined += i + "\n\n"
print(combined)
```

```python
text_file = open("combined.txt", "a+")
n = text_file.write(combined)
text_file.close()
```
